```python
import pandas as pd
import numpy as np
import spacy
import re

nlp = spacy.load('en_core_web_sm')
```

## Extract the sentences from the text and put the sentences into a list of pandas dataframes
#### Version 1: segment sentences from entries in set1.csv
#### Version 2: segment sentences from entries in set2.csv (all the leadership entries)
#### Version 3: segment sentences in student_responses.csv, then in mentor_responses.csv
#### Version 4: segment sentences in student_responses_up_to_f19.csv, then in mentor_responses_up_to_f19.csv 9-25-19
### Need to change it to work for both student and mentor, or at least choose which one to use

```python
ID = 0
all_sents = []
#all_commas = []

name_of_file = "mentor_responses_up_to_f19.csv"

with open("../data/" + name_of_file, "r", encoding="utf-8", errors="ignore") as in_file:
    for line in in_file:
        doc = nlp(line)
        sentences = list(doc.sents)

        sent_entry = pd.DataFrame()
        
        sent_entry['ID'] = [ID for i in range(len(sentences))]
        sent_entry['PID'] = [str(ID)+'_'+str(i) for i in range(len(sentences))]
        sent_entry['Text'] = [sentences[i] for i in range(len(sentences))]
        
        all_sents.append(sent_entry)
        ID += 1
```

## Concat the list of dataframes into one big table

```python
big_sents_table = pd.concat(all_sents)
```

## Write to a csv file

```python
big_sents_table.to_csv(path_or_buf="../data/mentor_responses_sentences_up_to_f19.csv", index=False)
```

## SHUFFLE pandas sample
